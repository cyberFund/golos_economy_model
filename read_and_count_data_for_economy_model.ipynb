```python
from pymongo import MongoClient
import numpy as np
import pandas as pd
```

```python
client = MongoClient("localhost:27018")
database = client["steemdb_1"]
collection_comment = database["comment"]
collection_account = database["account"]
```

Load comments from MongoDB

```python
cursor_comment = collection_comment.find({}, 
                                         {
                                            "reward_weight" : 1.0,
                                            "net_rshares" : 1.0,
                                            "net_votes" : 1.0,
                                            "total_payout_value" : 1.0,
                                            "root_comment" : 1.0,
                                            "last_payout" : 1.0,
                                            "depth" : 1.0
                                         })
```

```python
comment_df = pd.DataFrame(columns=['reward_weight', 
                                   'net_rshares',
                                   'total_payout_value',
                                   'net_votes', 
                                   'root_comment',
                                   'last_payout',
                                   'depth'])
for comment in cursor_comment:
    comment_df = comment_df.append(pd.DataFrame(comment, index=[len(comment_df)]))
comment_df.head()
```

```python
cursor_topic = collection_topic.find({
                                            "depth" : 0.0
                                         }, 
                                         {
                                            "reward_weight" : 1.0,
                                            "net_rshares" : 1.0,
                                            "net_votes" : 1.0,
                                            "total_payout_value" : 1.0,
                                            "root_comment" : 1.0,
                                            "last_payout" : 1.0
                                         })
```

```python
topic_df = pd.DataFrame(columns=['reward_weight', 
                                   'net_rshares',
                                   'total_payout_value',
                                   'net_votes', 
                                   'root_comment',
                                   'last_payout'])
for topic in cursor_topic:
    topic_df = topic_df.append(pd.DataFrame(topic, index=[len(comment_df)]))
topic_df.head()
```
